```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the MgMOF74_RAMAN_IR_Curve.csv file
MgMOF74 = np.genfromtxt('MgMOF-74_RAMAN_IR_Curve.csv', delimiter=',', skip_header=1).T
```

```python
fig, ax = plt.subplots(2, 3, figsize=(12, 5), gridspec_kw={'width_ratios': [2, 2, 1]})

for i in range(3):
    ax[0][i].plot(MgMOF74[0], MgMOF74[2]/max(MgMOF74[2]), '-', color='darkred', label='Simulated Raman')
    ax[1][i].plot(MgMOF74[0], MgMOF74[1]/max(MgMOF74[1]), color='darkred', label='Simulated FTIR')

for i in range(2):
    ax[i][0].legend(framealpha=0, loc='upper left', fontsize=12)
    ax[i][0].set_xlim(0, 1100)
    ax[i][0].set_ylim(0, 0.2)
    ax[i][0].set_ylabel('Intensity (a.u.)', fontsize=12, fontweight='bold')
    ax[i][0].set_yticks([])
    ax[i][0].set_xticks([0, 200, 400, 600, 800, 1000])
    ax[i][0].minorticks_on()
    ax[i][0].grid(axis='x', color='gray', linestyle='--', linewidth=0.7, alpha=0.2, zorder=2)
    ax[i][0].grid(axis='x', which='minor', color='grey', linestyle='--', alpha=0.2, linewidth=0.5, zorder=2)
    ax[i][0].text(0.95, 0.9, '5x', transform=ax[i, 0].transAxes, fontsize=12, fontweight='bold', rotation=0, va='center', ha='center')

for i in range(2):
    ax[i][1].set_xlim(1050, 1850)
    ax[i][1].set_ylim(0, 1.1)
    ax[i][1].set_yticks([])
    ax[i][1].minorticks_on()
    ax[i][1].grid(axis='x', color='gray', linestyle='--', linewidth=0.7, alpha=0.2, zorder=2)
    ax[i][1].grid(axis='x', which='minor', color='grey', linestyle='--', alpha=0.2, linewidth=0.5, zorder=2)

for i in range(2):
    ax[i][2].set_xlim(2850, 3450)
    ax[i][2].set_ylim(0, 0.5)
    ax[i][2].set_yticks([])
    ax[i][2].minorticks_on()
    ax[i][2].grid(axis='x', color='gray', linestyle='--', linewidth=0.7, alpha=0.2, zorder=2)
    ax[i][2].grid(axis='x', which='minor', color='grey', linestyle='--', alpha=0.2, linewidth=0.5, zorder=2)
    ax[i][2].text(0.9, 0.9, '2x', transform=ax[i, 2].transAxes, fontsize=12, fontweight='bold', rotation=0, va='center', ha='center')

ax[1][1].set_xlabel('Wavenumber (cm$^{-1}$)', fontsize=12, fontweight='bold')

plt.tight_layout()

# Reduce the horizontal space between plots
plt.subplots_adjust(wspace=0.015)

plt.savefig('raman.png', dpi=300)
plt.savefig('raman.pdf', dpi=300)
plt.show()
```
